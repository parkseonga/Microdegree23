```python
# count nans in arrs(3000) of x_arr(509281)
import collections

def check_nan_in_x_arr(arr):
    count = 0
    check_nan = np.isnan(arr)
    # print(check_nan)
    # print(collections.Counter(check_nan)[True])
    for index, boolen in enumerate(check_nan):
        if boolen == True:
            count += 1
    return count, collections.Counter(check_nan)[True]

total_nan_counts = 0
arr_nan_counts = 0
some_nans = []
all_nans = []
# some_nans = {}
# all_nans = {}
for index, arr in enumerate(x_arr):
    # count, nans = check_nan_in_x_arr(arr)
    count, collections_count = check_nan_in_x_arr(arr)
    assert count == collections_count, "counting error! (%d - %d, %d)" %(index, count, collections_count)
    if count > 0:
        arr_nan_counts += 1
        if count == len(arr): # 3000
            all_nans.append((index, count, collections_count))
            # all_nans[index] = count
        else:
            some_nans.append((index, count, collections_count))
            # some_nans[index] = count
        # print(index, count)
        # print(nans)
    total_nan_counts += count

print("total nans in x_arr : ", total_nan_counts)
print("arr with nans : ", arr_nan_counts)
print("- some of 3000 are nans :", len(some_nans), some_nans)
print("- all of 3000 are nans :", len(all_nans), all_nans)
# print("- some of 3000 are nans :", len(some_nans.keys()), sum(some_nans.values()), some_nans)
# print("- all of 3000 are nans :", len(all_nans.keys()), sum(all_nans.values()), all_nans)

# total nans in x_arr :  18222051
# arr with nans :  6984
# - some of 3000 are nans : 1318 
# - all of 3000 are nans : 5666
```

### # process x_arr data (get rid of nan)

```python
# change x_arr nan -> mean of previous values
some_nans_index = list(map(lambda x : x[0], some_nans))
print(len(some_nans_index), some_nans_index)

all_nans_index = list(map(lambda x : x[0], all_nans))
print(len(all_nans_index), all_nans_index)

def process_all_nan_arr(arr):
    arr = np.zeros(len(arr))
    return arr

def process_some_nan_arr(arr):
    check_nan = np.isnan(arr)
    for index, boolen in enumerate(check_nan):
        if boolen == True:
            if index == 0:
                arr[index] = 0
            else:
                arr[index] = np.mean(arr[:index])

for index, arr in enumerate(x_arr):
    if index in all_nans_index:
        x_arr[index] = process_all_nan_arr(arr)
        assert np.count_nonzero(x_arr[index]) == 0, '(%d) need all zero for all_nans_index !' %(index)
    elif index in some_nans_index:
        process_some_nan_arr(arr)
        assert collections.Counter(np.isnan(x_arr[index]))[True] == 0, '(%d) there is nan in some_nans_index !' %(index)
```
